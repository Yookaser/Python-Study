### 2) User-Defined Functions
- Built-in 함수 외에 우리가 직접 함수를 만들 수 있음. 이를 user-defined functions라고 함.
- User-defined function은 다음과 형태를 가짐.

```python
# def functionsName(par1, par2, ...):
#     indented block of statements
#     return expression
```

- par1, par2는 variables(변수)이고 parameters(파라미터)라고도 함.
- if, while,for 명령문처럼 function header는 콜론으로 끝나야 함.
- 각 블록(block)에 있는 명령문은 동일한 공백의 들여쓰기를 사용해야 함.(일반적으로 공백 4칸을 사용)
- 함수에는 여러 개의 return 명령문이 존재하지만, 첫번째 return 명령문이 실행되면 함수 실행은 종료됨.
- arguments에서 parameters로 전달되는 세가지 방법
    - pass by position
    - pass by keyword
    - pass by default value
- pass by position 방법으로 arguments가 parameters로 전달되는 경우, arguments와 parameters가 순서대로 매칭이 이루어짐.
    - 첫번째 argument는 첫번째 parameter에 전달되고, 두번째 argument는 두번째 parameter에 전달됨.
    - argument가 expression(표현식)인 경우 표현식의 값이 parameter에 전달됨.
- 나머지 두 방법(pass by keyword, pass by default value)은 추후 논의 예정
- parameters와 return 명령문이 반드시 함수에 포함될 필요는 없음.

### 6) Boolean- and List-Valued Functions
- 지금까지는 함수에 반환되는 값은 숫자나 문자열이었음. 하지만 다른 데이터 타입의 값도 반환이 가능함.

- Boolean 값을 반환하는 함수 예시

```python
# Ex.6

def isVowelWord(word):
    word = word.upper()
    vowels = ('A', 'E', 'I', 'O', 'U')
    for vowel in vowels:
        if vowel not in word:
            return False
    return True

word = input("Enter a word: ")
if isVowelWord(word):
    print(word, "contains every vowel.")
else:
    print(word, "does not contain every vowel.")
```

```python
# Ex.7

def occurringVowels(word):
    word = word.upper()
    vowels = ('A', 'E', 'I', 'O', 'U')
    includedVowels = []
    for vowel in vowels:
        if (vowel in word) and (vowel not in includedVowels):
            includedVowels.append(vowel)
    return includedVowels

word = input("Enter a word: ")
listOfVowels = occurringVowels(word)
print("The following vowels occur in the word: ", end="")
stringOfVowels = " ".join(listOfVowels)
print(stringOfVowels)
```

### 9) Scope of Variables
- 함수 내에서 생성된 변수는 함수 안의 명령문에 의해 접근이 가능함.
- 즉, 함수 실행이 사라지면 변수나 그 값도 같이 사라지게 됨.
- 함수가 재호출되면 그 변수나 값도 다시 만들어짐.
- 이러한 경우 "The variable is local to the function"이라고 하고, 그 변수는 "local scope"을 가진다고 얘기함.
- 서로 다른 함수에 동일한 이름의 변수들이 있다고 할 때, 두 변수는 서로 아무런 관계가 없음. 완전히 서로 다른 변수라고 할 수 있음.

```python
# Ex.10

def main():
    x = 2
    print(str(x) + ": function main")
    trivial()
    print(str(x) + ": function main")
def trivial():
    x = 3
    print(str(x) + ": function trivial")
main()
```

### 10) Global Variables

- 함수 내에서만 사용되는 변수를 local variable이라고 한다면, 모든 함수에 동일하게 적용되는 변수를 global variable이라고 함.
- Global variable은 일반적으로 프로그램 맨 윗줄에 함수가 정의되기 전에 정의됨.
- 정의된 모든 함수들이 global variable을 사용할 수 있지만 그 값을 변경하지는 못함.
- global 명령문을 사용하면 해당 함수 내에서 global variable의 값을 변경할 수 있음.
- global 명령문이 사용된 함수에서만 해당 변수를 변경할 수 있고, global 명령문이 사용되지 않은 함수에서는 그 변수값을 변경할 수 없음.

```python
a = 20 # global variable

def func1():
    a = 10 # local variable
    print(a)
    
def func2():
    print(a)
    
func1()
func2()
```

```python
a = 20 # global variable

def func1():
    global a # global variable
    a = 10 # global variable
    print(a)

def func2():
    print(a)
    
func1()
func2()
```

```python
# Ex.12

x = 0   # global variable 정의

def main():
    print(str(x) + ": function main")
    trivial()
    print(str(x) + ": function main")
    
def trivial():
    global x   # 함수 내에서 global variable을 변경할 수 있도록 global 명령문을 사용
    x += 7
    print(str(x) + ": function trivial")
    
main()
```

### 11) Named Constants
- 프로그램의 규모가 큰 경우에는 일반적으로 global variable 사용은 매우 제한적으로 허용되고 있음.
- 왜냐하면, 프로그램 내에서 global variable이 어디에서 정의되었는지 쉽게 확인하기 어렵고, 에러를 발생시키기도 하기 때문임.
- 이러한 문제가 있기 때문에 프로그램 내에 특별한 상수를 도입하기도 함. 이를 named constant라고 함.
- 일반적으로 named constants는 다음과 같이 대문자로 구성을 함.
    - INTEREST_RATE = 0.04
    - MINIMUM_VOTING_AGE = 18
    - BOOK_TITLE = "Programming with Python"
- 위와 같이 정의된 named constants는 프로그래머들이 협업을 할 때 그 값을 변경하지 않도록 주지시킬 수 있음.
- 프로그램 맨윗줄에 named constants가 정의되면, 나중에 그 값을 수정해야할 필요가 있을 때 관리가 쉬워짐.

```python
INTEREST_RATE = 0.04
MINIMUM_VOTING_AGE = 18
BOOK_TITLE = "Programming with Python"

amountDeposited = 2000
age = 20

interestEarned = INTEREST_RATE * amountDeposited

if (age >= MINIMUM_VOTING_AGE):
    print("You are eligible to vote.")
print("The title of the book is", BOOK_TITLE + ".")
```
