```python
# 6.4
#2(자리 수 더하기 함수)
def main():
    digitSum(12345)
    
def digitSum(n):
    total = 0
    while n:
        total += n % 10
        n = n // 10
    print(total)
    
main()
```

```python
#4(n승 함수)
def main():
    print(power(5,3))
    
def power(m,n):
    if n == 1:
        return m
    else:
        return m*power(m, n-1) # m*m*power(m,n-2) --> m*m*m*power(m,n-3)
    
main()
```

```python
#6(소인수 분해 함수)
def factor(n):
    if n==1:
        return []
    b = 2
    
    # b가 n보다 작거나 같으면 while문이 계속 실행됨.
    while b <= n:
        # n이 b로 나누어떨어지면 while문이 계속 실행됨.
        while not n % b:
            return [b] + factor(n//b) # n//b값이 1이 될 때까지 실행됨.
        b += 1
        
# n=12라고 하면, 2 <= 12이므로,
# 첫번째 while문이 실행됨.
# 12%2=0이므로,
# 두번째 while문이 실행됨.
# 그리고 [2] + factor(6)이 실행됨.

# factor(6)은 
# 2 <= 6이므로,
# 첫번째 while문이 실행됨.
# 6%2=0이므로,
# 두번째 while문이 실행됨.
# [2] + factor(3)이 실행됨.

# factor(3)은 
# 2<=3 이므로,
# 첫번째 while문이 실행됨.
# 3%2 != 0 이므로,
# b는 1이 더해져서 3이 됨.
# 3%3=0이므로,
# 두번째 while문이 실행됨.
# [3] + factor(1)이 실행됨.

# factor(1)은 []을 반환함.
```

```python
#10(피보나치 수열 함수)
def main():
    num = int(input("Enter a positive integer: "))
    print("Fibonacci number: ", fibonacci(num))

def fibonacci(n):
    if n == 1:
        return 1
    elif n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
main()
```

```python
#12(대출관련? 함수)
def main():
    p = eval(input("Enter the principal: "))
    r = eval(input("Enter the annual rate of interest: "))
    pmt = eval(input("Enter the monthly payment: "))
    n = eval(input("Enter the number of monthly payments made: "))
    print("The amount still owed is ${0:,.2f}".format(balance(p, pmt, r, n)))

def balance(p, pmt, r, n):
    if n == 0:
        return p
    else:
        return (1+(r/1200))*balance(p, pmt, r, n-1) - pmt
    
main()
```

```python
#14(합계 함수)
def summationList(L):
    L1 = list(L[:-1])
    if len(L1) == 0:
        return L[0]
    else:
        return L[-1] + summationList(L1)   
```

```python
L = [1,2,3,4,5]
summationList(L)
```
