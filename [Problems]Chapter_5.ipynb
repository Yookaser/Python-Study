```python
# 16

coefficientOfRestitution = eval(input("Enter coefficient of restitution: "))
initialHeight = eval(input("Enter initial height in meters: "))

numberOfBounces = 1
bouncedHeight = initialHeight
traveledMeters = initialHeight

while bouncedHeight > 0.1:
    bouncedHeight = bouncedHeight*(coefficientOfRestitution)**numberOfBounces
    numberOfBounces += 1
    traveledMeters += 2*bouncedHeight
    
print("Number of bounces: ", numberOfBounces)
print("Meters traveled: {0:.2f}".format(traveledMeters))    
```
