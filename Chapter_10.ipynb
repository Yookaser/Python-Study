### 3) 텍스트 파일로부터 딕셔너리 만들기
- 특정 프로그램이 큰 용량의 딕셔너리를 포함하고 있을 때, 그 딕셔너리는 일반적으로 텍스트 파일과 같은 형태의 파일로부터 만들어짐.

```python
# Ex.2

def main():
    texteseDict = createDictionary("Programs/Ch5/Textese.txt")
    print("Enter a simple sentence in lowercase letters without")
    sentence = input("any punctuation: ")
    print()
    translate(sentence, texteseDict)
    
def createDictionary(fileName):
    infile = open(fileName, 'r')
    textList = [line.rstrip() for line in infile]
    infile.close()
    return dict([x.split(',') for x in textList]) # 'adieu,u' --> ['adieu', 'u']

def translate(sentence, texteseDict):
    words = sentence.split()
    for word in words:
        print(texteseDict.get(word, word) + " ", end="")
        
main()
```

### 4) 빈도표로 딕셔너리 사용하기

```python
# Ex.4

def main():
    listOfWords = formListOfWords("Programs/Ch5/Gettysburg.txt")
    freq = createFrequencyDictionary(listOfWords)
    displayWordCount(listOfWords, freq)
    displayMostCommonWords(freq)
    
def formListOfWords(fileName):
    infile = open(fileName, 'r')
    originalLine = infile.readline().lower()
    
    line = ""
    for ch in originalLine:
        if ('a' <= ch <= 'z') or (ch == " "):
            line += ch
    listOfWords = line.split()
    return listOfWords

def createFrequencyDictionary(listOfWords):
    freq = {}
    for word in listOfWords:
        freq[word] = 0
    for word in listOfWords:
        freq[word] = freq[word] + 1
    return freq

def displayWordCount(listOfWords, freq):
    print("The Gettysburg Address contains", len(listOfWords), "words")
    print("The Gettysburg Address contains", len(freq), "different words.")
    print()
    
def displayMostCommonWords(freq):
    print("The most common words and their frequencies are: ")
    listOfMostCommonWords = []
    for word in freq.keys():
        if freq[word] >= 6:
            listOfMostCommonWords.append((word, freq[word])) # [('and', 5), ('the', 6),...]
    listOfMostCommonWords.sort(key=lambda x: x[1], reverse=True)
    for item in listOfMostCommonWords:
        print("   ", item[0] + ':', item[1])

main()
```

### 7) 딕셔너리로부터 정렬된 데이터 추출하기
- 딕셔너리 내 데이터는 순서를 갖지 않기 때문에 sort() 메소드를 사용할 수 없음.
- 하지만, 딕셔너리 내 아이템(key:value 쌍)은 리스트 안에 내포된 튜플 형태로 다음과 같은 명령문으로 변경할 수 있음.
    - sorted(dict1.items(), key=f, reverse = BooleanValue)

```python
# Ex.6

import pickle

def main():
    nations = getDictionary("UNdict.dat")
    continent = input("Enter the name of a continent other than Antarctica: ")
    continentDict = constructContinentNations(nations, continent)
    displaySortedResults(continentDict)
    
def getDictionary(fileName):
    infile = open(fileName, 'rb')
    nations = pickle.load(infile)
    infile.close()
    return nations

def constructContinentNations(nations, continent):
    continentDict = {}
    for nation in nations:
        if nations[nation]["cont"] == continent: # 입력한 대륙명 관련 데이터만 추출
            continentDict[nation] = nations[nation]
    return continentDict

def displaySortedResults(dictionaryName):
    continentList = sorted(dictionaryName.items(),
                          key=lambda k:k[1]["popl"], reverse=True)
    for k in continentList:
        print("  {0:s}: {1:,.2f}".format(k[0], k[1]["popl"]))
        
main()
```

```python
# Ex.7

import pickle

def main():
    presDict = createDictFromBinaryFile("USpresStatesDict.dat")
    state = getState(presDict)
    displayOutput(state, presDict)
    
def createDictFromBinaryFile(fileName):
    infile = open(fileName, 'rb')
    dictionaryName = pickle.load(infile)
    infile.close()
    return dictionaryName

def getState(dictName):
    state = input("Enter the name of a state: ")
    if state in dictName.values(): # 딕셔너리 value에 입력한 주의 이름이 존재하면 해당 주의 값을 반환
        return state
    else: # 딕셔너리 value에 입력한 주의 이름이 존재하지 않으면 아래 문구를 반환
        return "There are no presidents from " + state + '.'
    
def displayOutput(state, dictName):
    if state.startswith("There"): # 딕셔너리 value에 입력한 주의 이름이 존재하지 않는 경우
        print(state) # "There are no presidents from..." 문구를 그대로 출력
    else:
        print("Presidents from", state + ':')
        for pres in sorted(dictName): # sorted(d1)은 d1의 key값이 정렬된 형태의 리스트 데이터를 반환
            if dictName[pres] == state: # 
                print("  " + pres[1] + "  " + pres[0])
                
main()
```

### 9) 딕셔너리 컴프리헨션
